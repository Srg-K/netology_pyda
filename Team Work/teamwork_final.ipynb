```python
import pandas as pd
import numpy as np
import pandas_profiling
import matplotlib.pyplot as plt
import scipy.stats as sts
import seaborn as sns
import statistics
```

```python
df = pd.read_csv("AIC_final.csv")
df.head()
```

```python
def qty(row):
    x = int(row["Quantity"])
    return x

df["qty"] = df.apply(qty, axis=1)
df.drop(['Quantity'], axis=1, inplace=True)
df.head()
```

```python
def revenue(row):
    x = int(row.UnitPrice.replace(',',''))
    revenue = row['qty'] * x / 100
    return revenue
df["revenue"] = df.apply(revenue, axis=1)
df.head()
```

```python
def price(row):
    x = int(row.UnitPrice.replace(',',''))
    pp = x / 100
    return pp
df["price"] = df.apply(price, axis=1)
df.head()
```

```python
def dateformat(row):
    x = row.InvoiceDate.split()[0]
    return x

df['date'] = df.apply(dateformat, axis=1)
df.head()
```

```python
df.drop(['InvoiceDate'], axis=1, inplace=True)
```

```python
df = df.dropna()
df.head()
```

```python
def uid(row):
    x = int(row.CustomerID)
    return x

df['uid'] = df.apply(uid, axis=1)
df.head()
```

```python
df.drop(['CustomerID'], axis=1, inplace=True)
```

```python
item_price = df[['StockCode', 'price']]
item_price = item_price.groupby('StockCode').mean().sort_values(by='price', ascending=False)

item_revenue = df[['StockCode', 'revenue']]
item_revenue = item_revenue.groupby('StockCode').sum().sort_values(by='revenue', ascending=False)

price_segments = item_price.merge(item_revenue, on='StockCode')
#price_segments['revenue'].quantile(0.75)
```

```python
def psegment(row):
    if row.price < 250:
        return 'low'
    elif row.price < 500:
        return 'mid'
    else:
        return 'high'
    
price_segments['segment'] = price_segments.apply(psegment, axis=1)
price_segments = price_segments.groupby('segment').sum()
price_segments
```
